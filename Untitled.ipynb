```python
import requests as req
from bs4 import BeautifulSoup as bs4
import pandas as pd
from pandas.plotting import register_matplotlib_converters
!pip install lxml
!pip install html5lib
from matplotlib import pyplot as plt
import datetime as dt
```

```python
coinmarketcap_url = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20191208"
r = req.get(coinmarketcap_url)
root = bs4(r.content, 'html.parser')
```

```python
thead = root.find('thead')
tbody = root.find('tbody')
table = '<table>' + str(thead) + str(tbody) + '</table>'
```

```python
coinmarketcap = pd.read_html(str(table))[0]
```

```python
datahub = pd.read_csv('bitcoin_csv.csv')
```

```python
coinmarketcap.head()
```

```python
datahub.head()
```

```python
coinmarketcap_dates = []
for i, row in coinmarketcap.iterrows():
    coinmarketcap_dates.append(dt.datetime.strptime(row['Date'], '%b %d, %Y').date())
coinmarketcap['Date'] = coinmarketcap_dates
```

```python
datahub_dates = []
for i, row in datahub.iterrows():
    datahub_dates.append(dt.datetime.strptime(row['date'], '%Y-%m-%d').date())
datahub['date'] = datahub_dates
```

```python
coinmarketcap = coinmarketcap.rename(columns={'Date':'date'})
```

```python
coinmarketcap = coinmarketcap[coinmarketcap['date'] <= dt.date(2018, 11, 22)]
coinmarketcap = coinmarketcap.sort_values(by='date')
```

```python
datahub = datahub[datahub['date'] >= dt.date(2013, 4, 28)]
```

```python
data = pd.merge(coinmarketcap, datahub, on='date', how='inner')
```
