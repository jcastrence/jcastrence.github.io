```python
# Data Cleaning

# What is data cleaning?

# How will we clean data?
```

```python
# https://coinmetrics.io/community-data-dictionary/
# Cleaning Coinmetrics
def clean_coinmetrics(df):
    dates = []
    for i, row in df.iterrows():
        dates.append(dt.datetime.strptime(row['date'], '%Y-%m-%d').date())
    df['date'] = dates
    df = df.dropna(subset=['CapMrktCurUSD'])
    df = df.reset_index(drop=True)
    return df

# Bitcoin
btc = clean_coinmetrics(btc)

# Ethereum
eth = clean_coinmetrics(eth)

# Ripple
xrp = clean_coinmetrics(xrp)

# Bitcoin Cash
bch = clean_coinmetrics(bch)

#Litecoin
ltc = clean_coinmetrics(ltc)
```

```python
# Cleaning CoinMarketCap
coinmarketcap_dates = []
for i, row in coinmarketcap.iterrows():
    coinmarketcap_dates.append(dt.datetime.strptime(row['Date'], '%b %d, %Y').date())
coinmarketcap['Date'] = coinmarketcap_dates
coinmarketcap = coinmarketcap.rename(columns={'Date':'date', 'Open*':'open', 'High':'high', 'Low':'low', \
                                              'Close**':'close', 'Volume':'volume', 'Market Cap':'marketCap'})
coinmarketcap = coinmarketcap.sort_values(by='date')
coinmarketcap = coinmarketcap.reset_index(drop=True)
```

```python
# Cleaning Google Trends
googletrends = googletrends.drop('Day')
googletrends = googletrends.reset_index()
googletrends = googletrends.rename(columns={'index':'date', 'Category: All categories':'searchPopularity'})
googletrends_dates = []
googletrends_search = []
for i, row in googletrends.iterrows():
    googletrends_dates.append(dt.datetime.strptime(row['date'], '%Y-%m-%d').date())
    googletrends_search.append(int(row['searchPopularity']))
googletrends['date'] = googletrends_dates
googletrends['searchPopularity'] = googletrends_search
```

```python
# Calculate 14 day RSI and 12-26-9 MACD
indicators = coinmarketcap[['date', 'close']]
```

```python
indicators.head()
```

```python
change = []
for i, row in indicators.iterrows():
    x = row.close - indicators.loc[i-1, 'close'] if i > 0 else np.nan
    change.append(x)
indicators['change'] = change
indicators = indicators.dropna()
indicators = indicators.reset_index(drop=True)
indicators['upMove'] = [x if x >= 0 else 0 for x in indicators['change']]
indicators['downMove'] = [abs(x) if x < 0 else 0 for x in indicators['change']]
```

```python
avgUpMove = []
for i, row in indicators.iterrows():
    x = np.nan
    if i == 13:
        x = indicators[0:14]['upMove'].mean()
    elif i > 13:
        x = (avgUpMove[i-1] * 13 + row['upMove']) / 14
    avgUpMove.append(x)
indicators['avgUpMove'] = avgUpMove
```

```python
avgDownMove = []
for i, row in indicators.iterrows():
    x = np.nan
    if i == 13:
        x = indicators[0:14]['downMove'].mean()
    elif i > 13:
        x = (avgDownMove[i-1] * 13 + row['downMove']) / 14
    avgDownMove.append(x)
indicators['avgDownMove'] = avgDownMove
```

```python
indicators['RS'] = indicators['avgUpMove'] / indicators['avgDownMove']
indicators['RSI'] = 100 - (100 / (indicators['RS'] + 1))
```

```python
# Calculate 12 day EMA
EMA12 = []
for i, row in indicators.iterrows():
    x = np.nan
    if i == 11:
        x = indicators[0:12]['close'].mean()
    elif i > 11:
        x = (row['close'] - EMA12[i-1]) * (2 / 13) + EMA12[i-1]
    EMA12.append(x)
indicators['EMA12'] = EMA12
```

```python
# Calculate 26 day EMA
EMA26 = []
for i, row in indicators.iterrows():
    x = np.nan
    if i == 25:
        x = indicators[0:26]['close'].mean()
    elif i > 25:
        x = (row['close'] - EMA26[i-1]) * (2 / 27) + EMA26[i-1]
    EMA26.append(x)
indicators['EMA26'] = EMA26
```

```python
# Calculate MACD Line
indicators['MACD'] = indicators['EMA26'] - indicators['EMA12']
indicators = indicators.dropna()
indicators = indicators.reset_index(drop=True)
```

```python
# Calculate Signal Line (9 Day EMA for MACD Line)
signal = []
for i, row in indicators.iterrows():
    x = np.nan
    if i == 8:
        x = indicators[0:9]['MACD'].mean()
    elif i > 8:
        x = (row['MACD'] - signal[i-1]) * (2 / 10) + signal[i-1]
    signal.append(x)
indicators['signal'] = signal
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0.5, 0.8, 0.8])
ax2 = fig.add_axes([0, 0.3, 0.8, 0.3])

ax1.plot(indicators.date, indicators.close)
ax2.plot(indicators.date, indicators.RSI)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0.5, 0.8, 0.8])
ax2 = fig.add_axes([0, 0.3, 0.8, 0.3])

macd = ['MACD', 'signal']

ax1.plot(indicators.date, indicators.close)
for col in macd:
    ax2.plot(indicators.date, indicators[col])
ax2.legend(macd)
```
